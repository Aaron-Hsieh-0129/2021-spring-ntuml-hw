```python
"""
Roberta-AdamW
"""

import numpy as np
import pandas as pd

# =========== read file ========================
data_root='/content/drive/MyDrive/ML/hw7/result/roberta/'

def read_data(path, n):
    tmpData = np.asarray(pd.read_csv(data_root + f"result{n}.csv"))[:, 1]
    return tmpData

data_num = 7
data = np.array([])
for i in range(1, data_num+1, 1):
    data = np.append(data, read_data(f"result{i}.csv", i))

data = data.reshape(data_num, int(len(data) / data_num))
data = data.transpose()

# =========== weight ========================
model_accuracy = np.array([0.898, 0.892, 0.895, 0.898, 0.894, 0.894, 0.896])
std = np.std(model_accuracy, ddof=0)
z = (model_accuracy - np.mean(model_accuracy)) / std
weight = 1.25 ** z


# =========== voting ========================
final = []
for i in range(data.shape[0]):
    check = {}
    for j in range(data.shape[1]):
        if check.get(data[i][j], "None") == "None":
            check[data[i][j]] = weight[j]
        else:
            check[data[i][j]] = check[data[i][j]] + weight[j]

    final.append(max(check, key=check.get))

final = np.asarray(final)
```

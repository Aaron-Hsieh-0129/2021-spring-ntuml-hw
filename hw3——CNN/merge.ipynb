```python
import numpy as np
import pandas as pd

data_root='/content/drive/MyDrive/Colab Notebooks/ML/hw3_all/'

data1 = np.asarray(pd.read_csv(data_root + "predict1.csv"))[:, 1]
data2 = np.asarray(pd.read_csv(data_root + "predict2.csv"))[:, 1]
data3 = np.asarray(pd.read_csv(data_root + "predict3.csv"))[:, 1]
data4 = np.asarray(pd.read_csv(data_root + "predict4.csv"))[:, 1]
data5 = np.asarray(pd.read_csv(data_root + "predict5.csv"))[:, 1]
data6 = np.asarray(pd.read_csv(data_root + "predict6.csv"))[:, 1]
data7 = np.asarray(pd.read_csv(data_root + "predict7.csv"))[:, 1]


data = np.array([])
data = np.append(data, data1)
data = np.append(data, data2)
data = np.append(data, data3)
data = np.append(data, data4)
data = np.append(data, data5)
data = np.append(data, data6)
data = np.append(data, data7)


data = data.reshape(7, len(data1))
data = data.transpose()
data = data.astype('int')
final = np.zeros(len(data1))

model_accuracy = np.array([0.78375, 0.71445, 0.73118, 0.70131, 0.70669, 0.68876, 0.68757])
std = np.std(model_accuracy, ddof=0)
z = (model_accuracy - np.mean(model_accuracy)) / std
weight = 1.25 ** z
print(weight)

for i in range(len(data1)):
    check = np.zeros(11)
    for j in range(7):
        if j == 0:
            check[data[i][j]] += weight[j]
        elif j == 1:    
            check[data[i][j]] += weight[j]
        elif j == 2:
            check[data[i][j]] += weight[j]    
                                

    final[i] = np.argmax(check)


final = final.astype('int')
# Save predictions into the file.
with open("predict.csv", "w") as f:

    # The first row must be "Id, Category"
    f.write("Id,Category\n")

    # For the rest of the rows, each image id corresponds to a predicted class.
    for i, pred in  enumerate(final):
         f.write(f"{i},{pred}\n")
```
